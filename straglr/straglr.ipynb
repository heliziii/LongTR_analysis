```python
import pandas as pd
import numpy as np
from cyvcf2 import VCF
```

```python
###### Read LongTR output on HG002 PacBio #######
calls = []
vcf = VCF(f"/expanse/projects/gymreklab/helia/HipSTR_LR/tests/whole_genome/output/HG002_wg_HiFi_HipSTR_long_chr21.vcf.gz")
for variant in vcf:
    if variant.format('GB')[0] == ".":
        continue
    calls.append([variant.CHROM, variant.POS, variant.INFO['START'], variant.INFO['END'], variant.INFO['PERIOD'],
                  variant.REF, variant.ID, variant.gt_bases[0].split("|"), variant.INFO['DP'],
                  variant.format('GB')[0].split("|")])
PacBio_data = pd.DataFrame(calls, 
    columns = ["chrom","pos", "start", "end", "period", "ref", "ID", "gbs", "dp", 'gb'])

non_homo_21 = PacBio_data[(PacBio_data['chrom'] == 'chr21')]
non_homo_21
```

```python
straglr_pacbio_data = pd.read_csv("output/pacbio_HG002_chr21.bed", sep = "\t")
print(len(straglr_pacbio_data))
PacBio = pd.merge(straglr_pacbio_data, non_homo_21, on = ['start','end'], how = "inner")
PacBio
```

```python
def compare(gb_longtr, allele1, allele2, start, end):
    if allele1 == "-":
        return np.nan
    if gb_longtr[0] == ".":
        return np.nan
    start = int(start)
    end = int(end)
    diff = 2
    gb_straglr = []
    gb1 = float(allele1) - (end - start + 1)
    gb2 = "."
    if allele2 != "-":
        gb2 = float(allele2) - (end - start + 1)
    if gb2 != ".":
        gb_straglr = [gb1,gb2]
    else:
        gb_straglr = [gb1,gb1]
    gb_longtr = sorted([int(x) for x in gb_longtr])
    gb_straglr = sorted([int(x) for x in gb_straglr])
    if abs(gb_longtr[0] - gb_straglr[0]) < diff and abs(gb_longtr[1] - gb_straglr[1]) < diff:
        return 1
    if abs(gb_longtr[0] - gb_straglr[0]) < diff or abs(gb_longtr[1] - gb_straglr[1]) < diff:
        return 0.5
    if abs(gb_longtr[0] - gb_straglr[1]) < diff or abs(gb_longtr[0] - gb_straglr[1]) < diff:
        return 0.5
    return 0

PacBio['compare'] = PacBio.apply(lambda row: compare(row['gb'], 
                                                   row['allele1:size'], row['allele2:size'],
                                                    row['start'], row['end']), axis = 1)
```
