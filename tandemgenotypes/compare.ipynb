```python
import pandas as pd
import numpy as np
from cyvcf2 import VCF
```

```python
###### Read LongTR output on HG002 PacBio #######
calls = []
vcf = VCF(f"/expanse/projects/gymreklab/helia/HipSTR_LR/TRGT/HG002_wg_hipstr_adotto_apr2.vcf.gz")
for variant in vcf:
    if variant.format('GB')[0] == ".":
        continue
    calls.append([variant.CHROM, variant.POS, variant.INFO['START'], variant.INFO['END'], variant.INFO['PERIOD'],
                  variant.REF, variant.ID, variant.gt_bases[0].split("|"), variant.INFO['DP'],
                  variant.format('GB')[0].split("|")])
PacBio_data = pd.DataFrame(calls, 
    columns = ["chrom","pos", "start", "end", "period", "ref", "ID", "gbs", "dp", 'gb'])

non_homo_21 = PacBio_data[(PacBio_data['chrom'] == 'chr21')]
non_homo_21
```

```python
TG_pacbio_data = pd.read_csv("HG002_chr21_TG_output.txt", sep = "\t", skiprows=1, header=None)
print(len(TG_pacbio_data))
PacBio = pd.merge(TG_pacbio_data, non_homo_21, right_on = ['start','end'], left_on=[1,2], how = "inner")
PacBio
```

```python
def compare(gb_longtr, allele1, allele2, start, end, period):
    if allele1 == ".":
        return np.nan
    if gb_longtr[0] == ".":
        return np.nan
    start = int(start)
    end = int(end)
    diff = period
    gb_straglr = []
    gb1 = float(allele1) * period
    gb2 = "."
    if allele2 != ".":
        gb2 = float(allele2) * period
    if gb2 != ".":
        gb_tg = [gb1,gb2]
    else:
        gb_tg = [gb1,gb1]
    gb_longtr = sorted([int(x) for x in gb_longtr])
    gb_tg = sorted([int(x) for x in gb_tg])
    if abs(gb_longtr[0] - gb_tg[0]) < diff and abs(gb_longtr[1] - gb_tg[1]) < diff:
        return 1
    if abs(gb_longtr[0] - gb_tg[0]) < diff or abs(gb_longtr[1] - gb_tg[1]) < diff:
        return 0.5
    if abs(gb_longtr[0] - gb_tg[1]) < diff or abs(gb_longtr[0] - gb_tg[1]) < diff:
        return 0.5
    return 0

PacBio['compare'] = PacBio.apply(lambda row: compare(row['gb'], 
                                                   row[6], row[7],
                                                    row['start'], row['end'], row['period']), axis = 1)
PacBio
```
